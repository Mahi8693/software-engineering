```python
# Version 1: Read single input from file

try:
    with open("single_input.txt", "r") as file:
        number = int(file.read().strip())
        print("Number:", number)
except FileNotFoundError:
    print("Error: File not found.")
except ValueError:
    print("Error: Invalid input. Expected an integer.")
```

```python
# Version 2: Read multiple inputs from file

try:
    with open("multiple_inputs.txt", "r") as file:
        numbers = [int(line.strip()) for line in file]

    total = sum(numbers)
    average = total / len(numbers)

    print("Numbers:", numbers)
    print("Sum:", total)
    print("Average:", average)

except FileNotFoundError:
    print("Error: File not found.")
except ValueError:
    print("Error: File contains non-numeric data.")
except ZeroDivisionError:
    print("Error: No numbers found in file.")
```

```python
# Version 3: Robust error handling

def read_numbers(filename):
    try:
        with open(filename, "r") as file:
            numbers = [int(line.strip()) for line in file if line.strip()]
            if not numbers:
                raise ValueError("File is empty.")
            return numbers
    except Exception as e:
        print("Error:", e)
        return []

nums = read_numbers("multiple_inputs.txt")
if nums:
    print("Sum:", sum(nums))
    print("Average:", sum(nums)/len(nums))
```
